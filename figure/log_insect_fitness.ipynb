```r
library(tidyverse)
```

```r
data <- read.csv("https://raw.githubusercontent.com/BIO465/2019_Winter_Nematode_Bacteria/master/data/10_13genes.csv")
```

```r
data$InsectFitness[data$InsectFitness == "#DIV/0!"] <- 0
data$InsectFitness <- log(as.numeric(as.character(data$InsectFitness)))
```

```r
ggplot(data, aes(x=as.numeric(as.character(data$InsectFitness)))) +
  geom_histogram(bins=100) +
  geom_vline(xintercept = 5.4, color="red", linetype=2) +
  theme_bw() +
  theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank()) +
  xlab("log(Insect Fitness)") +
  ylab("Frequency")
```
